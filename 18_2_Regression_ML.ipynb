# Multivariate Regression

```
# 앞의 강의와 동일한 데이터
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/rasbt/'
                 'python-machine-learning-book-3rd-edition/'
                 'master/ch10/housing.data.txt', sep='\s+', header=None ) # \s+는 space 나 tab 이 섞여있을 때

df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 
              'NOX', 'RM', 'AGE', 'DIS', 'RAD', 
              'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.head()

# 1. CRIM      per capita crime rate by town
# 2. ZN        proportion of residential land zoned for lots over 
#                  25,000 sq.ft.
# 3. INDUS     proportion of non-retail business acres per town
# 4. CHAS      Charles River dummy variable (= 1 if tract bounds 
#                  river; 0 otherwise)
# 5. NOX       nitric oxides concentration (parts per 10 million)
# 6. RM        average number of rooms per dwelling
# 7. AGE       proportion of owner-occupied units built prior to 1940
# 8. DIS       weighted distances to five Boston employment centres
# 9. RAD       index of accessibility to radial highways
# 10. TAX      full-value property-tax rate per $10,000
# 11. PTRATIO  pupil-teacher ratio by town
# 12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
#                  by town
# 13. LSTAT    % lower status of the population
# 14. MEDV     Median value of owner-occupied homes in $1000s
```

```

X = df[['RM', 'AGE', 'DIS']].values   # values를 사용하면 데이터프레임의 데이터만 가져온다 (인덱스명과 컬럼명 제외, 판다스 데이터 프레임은 데이터를 numpy array로 반환)
y = df.MEDV.values
```

```
print(X.shape)
print(y.shape)
```

```
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X,y)
print(mlr.coef_)
print(mlr.intercept_)
print(mlr.score(X,y))
```

# 모델 성능 평가
테스트 셋을 통해 성능을 확인해봅시다

```
from sklearn.model_selection import train_test_split
X = df.iloc[:,:-1].values
y = df.MEDV.values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

```
# testset의 resudual을 구해봅시다
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(y_test_pred, y_test_pred-y_test)
plt.title('residuals') # 타이틀
plt.xlabel('y_test_pred') # y축 제목
plt.ylabel('residuals') # x축 제목
plt.hlines(y=0, xmin=y_test_pred.min(), xmax=y_test_pred.max())  # 수평선 그리기
plt.show()
```

```
# trainset의 resudual을 구해서 비교해봅시다
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(y_train_pred, y_train_pred-y_train)
plt.title('residuals') # 타이틀
plt.xlabel('y_train_pred') # y축 제목
plt.ylabel('residuals') # x축 제목
plt.hlines(y=0, xmin=y_train_pred.min(), xmax=y_train_pred.max())  # 수평선 그리기
plt.show()
```

```
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
print(f'train MSE: {mse_train:.2f}, test MSE: {mse_test:.2f}')
```

# randomforest로 regression 모델 만들기

위의 모델을 randomforest로 만들어봅시다

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html#sklearn.ensemble.RandomForestRegressor

```
from sklearn.ensemble import RandomForestRegressor  # randomforest regressor 불러오기

forest_rgr = RandomForestRegressor(max_depth=4, random_state=0)  # 인스턴스 만들기
```

```
forest_rgr.fit(X,y)  # 위와 동일한 데이터 호출
```

```
imp = forest_rgr.feature_importances_
```

```
imp_df = pd.Series(imp, df.columns[:-1])
```

```
imp_df.sort_values(ascending = False).plot.bar()
```

성능을 확인해봅시다

```
y_train_pred = forest_rgr.predict(X_train)
y_test_pred = forest_rgr.predict(X_test)
```

```
# testset의 resudual을 구해봅시다
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(y_test_pred, y_test_pred - y_test)
plt.title('residuals') # 타이틀
plt.xlabel('y_test_pred') # y축 제목
plt.ylabel('residuals') # x축 제목
plt.hlines(y=0, xmin=y_test_pred.min(), xmax=y_test_pred.max())  # 수평선 그리기
plt.show()
```

```
from sklearn.metrics import mean_squared_error
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
print(f'train MSE: {mse_train:.2f}, test MSE: {mse_test:.2f}')
```
