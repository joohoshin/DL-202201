함수의 입력값을 몇개를 받을 지 모를때 * 표시를 사용하면 
입력값을 리스트 형태로 받을 수 있다.

```python
def print_names(*args):
  for i in args:
    print(i)
```

```python
print_names('a','b','c')
```

**를 사용하면 dictionary 형태로 입력을 받을 수 있다

```python
def setting(**kwargs):
  for key, value in kwargs.items():
    print(key, value)
```

```python
setting(epoch = 10, size = 20, batch=30)
```
